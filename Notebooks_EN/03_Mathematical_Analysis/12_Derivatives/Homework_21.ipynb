## 21. Derivatives

1. Compute derivatives of functions:

   - $y(x) = -3x + 3$
   $$
   \frac{dy}{dx} = -3
   $$

   - $y(x) = \pi x + \sin(1)$
   $$
   \frac{dy}{dx} = \pi
   $$

   - $y(x) = 4 + \sin(2)$
   $$
   \frac{dy}{dx} = 0
   $$

   - $y(x) = 2x^3 - 3x^2 + 8x - 9$
   $$
   \frac{dy}{dx} = 6x^2 - 6x + 8
   $$

   - $y(x) = 6x^{1/3}$
   $$
   \frac{dy}{dx} = 2x^{-2/3}
   $$

   - $y(x) = \sqrt{x}$
   $$
   \frac{dy}{dx} = \frac{1}{2\sqrt{x}}
   $$

   - $y(x) = \cos(x) + \sin(x)$
   $$
   \frac{dy}{dx} = -\sin(x) + \cos(x)
   $$

   - $y(x) = 2\sin(x) \cos(x)$
   Using the product rule:
   $$
   \frac{dy}{dx} = 2[\cos(x) \cos(x) - \sin(x) \sin(x)] = 2\cos(2x)
   $$

   - $y(x) = x\sin(x)$
   Using the product rule:
   $$
   \frac{dy}{dx} = \sin(x) + x\cos(x)
   $$

   - $y(x) = (x+1)(x+1)$
   Expanding and differentiating:
   $$
   \frac{dy}{dx} = 2(x+1)
   $$

   - $y(x) = \frac{x}{x+1}$
   Using the quotient rule:
   $$
   \frac{dy}{dx} = \frac{(x+1)(1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2}
   $$

   - $y(x) = (x+1)\exp(x)$
   Using the product rule:
   $$
   \frac{dy}{dx} = \exp(x) + (x+1)\exp(x) = (x+2)\exp(x)
   $$

   - $y(x) = \sin(x^2)$
   Using the chain rule:
   $$
   \frac{dy}{dx} = 2x\cos(x^2)
   $$

   - $y(x) = \exp(-2x)$
   $$
   \frac{dy}{dx} = -2\exp(-2x)
   $$

   - $y(x) = \frac{1}{\sin(x+1)}$
   Using the chain rule:
   $$
   \frac{dy}{dx} = -\frac{\cos(x+1)}{\sin^2(x+1)}
   $$

   - $y(x) = \sqrt{2x+1}$
   Using the chain rule:
   $$
   \frac{dy}{dx} = \frac{1}{\sqrt{2x+1}} \cdot 2 = \frac{2}{\sqrt{2x+1}}
   $$

2. Prove:
   - $\frac{d}{dx} (\ln(\sin(x))) = \cot(x)$

   Using the chain rule:
   $$
   \frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x)
   $$

3. For $f(x) = \cos(x)$, verify that $f''(x) = -f(x)$.

   First, compute the first derivative:
   $$
   f'(x) = -\sin(x)
   $$
   Now, compute the second derivative:
   $$
   f''(x) = -\cos(x)
   $$
   Therefore, $f''(x) = -f(x)$, as required.

4. Using de l'Hospital's Rule, find the improper limits:

   - $\displaystyle \lim_{x \to 0} \frac{\sin{x}}{x}$

   Applying de l'Hospital's Rule:
   $$
   \lim_{x \to 0} \frac{\sin{x}}{x} = \lim_{x \to 0} \frac{\cos{x}}{1} = 1
   $$

   - $\displaystyle \lim_{x \to \infty} \frac{\ln x}{x}$

   Applying de l'Hospital's Rule:
   $$
   \lim_{x \to \infty} \frac{\ln x}{x} = \lim_{x \to \infty} \frac{1/x}{1} = 0
   $$

   - $\displaystyle \lim_{x \to \infty} \frac{\exp(x)}{x}$

   Applying de l'Hospital's Rule:
   $$
   \lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\exp(x)}{1} = \infty
   $$

5. In physics, the position of a particle is given by $x(t) = 3t^2 - 6t + 1$. Find the velocity $V(t)=x'(t)$ and acceleration $a(t)=V'(t)=x''(t)$ of the particle at time $t = 2$.

   The velocity is the first derivative of position:
   $$
   V(t) = \frac{d}{dt} (3t^2 - 6t + 1) = 6t - 6
   $$
   At $t = 2$:
   $$
   V(2) = 6(2) - 6 = 6
   $$

   The acceleration is the derivative of velocity:
   $$
   a(t) = \frac{d}{dt} (6t - 6) = 6
   $$

   Therefore, the velocity at $t = 2$ is $V(2) = 6$ and the acceleration is $a(2) = 6$.
