## Problem

Let \( f(x) = e^x \) and \( g(x) = \ln(x) \). Check \( f(g(x)) \) and \( g(f(x)) \). What do you notice?

### Solution

1. **Find \( f(g(x)) \):**

   We need to substitute \( g(x) = \ln(x) \) into \( f(x) \):
   \[
   f(g(x)) = f(\ln(x)) = e^{\ln(x)}
   \]
   Using the property of logarithms and exponentials that \( e^{\ln(x)} = x \), we get:
   \[
   f(g(x)) = x
   \]

2. **Find \( g(f(x)) \):**

   We need to substitute \( f(x) = e^x \) into \( g(x) \):
   \[
   g(f(x)) = g(e^x) = \ln(e^x)
   \]
   Using the property that \( \ln(e^x) = x \), we get:
   \[
   g(f(x)) = x
   \]

### What do you notice?

Both \( f(g(x)) \) and \( g(f(x)) \) simplify to \( x \). This is due to the inverse relationship between the exponential function \( e^x \) and the natural logarithm \( \ln(x) \). Specifically:

- \( e^x \) is the inverse function of \( \ln(x) \), and vice versa.
- Therefore, applying one function to the other results in the identity function, i.e., \( f(g(x)) = g(f(x)) = x \).

### Final Answers

- \( f(g(x)) = x \)
- \( g(f(x)) = x \)


## Problem

We have the function \( f = \{(1, 7), (2, 9), (3, 11)\} \). Find the inverse function \( f^{-1} \).

### Solution

The inverse function \( f^{-1} \) is obtained by swapping the elements of each ordered pair in \( f \). That is, for each pair \( (a, b) \) in \( f \), we write the pair \( (b, a) \) in \( f^{-1} \).

Given:
\[
f = \{(1, 7), (2, 9), (3, 11)\}
\]

We swap the first and second elements of each pair:
\[
f^{-1} = \{(7, 1), (9, 2), (11, 3)\}
\]

### Final Answer

The inverse function is:
\[
f^{-1} = \{(7, 1), (9, 2), (11, 3)\}
\]


## Problem

We have the function \( f = \{(1, 7), (2, 7), (3, 11)\} \). Find the inverse function \( f^{-1} \).

### Solution

The inverse function \( f^{-1} \) is obtained by swapping the elements of each ordered pair in \( f \). That is, for each pair \( (a, b) \) in \( f \), we write the pair \( (b, a) \) in \( f^{-1} \).

However, in this case, we notice that the pair \( (1, 7) \) and \( (2, 7) \) both map to the same value \( 7 \), which means that \( f \) is not a one-to-one function. Since the inverse function requires that each output value is uniquely mapped to an input value, the function \( f \) does not have an inverse.

### Conclusion

The function \( f = \{(1, 7), (2, 7), (3, 11)\} \) does not have an inverse, because it is not one-to-one (i.e., \( 7 \) corresponds to both \( 1 \) and \( 2 \)).
