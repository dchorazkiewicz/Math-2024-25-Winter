### Problem 6:
The profit function is given as \( P(u) = -2u^2 + 50u - 300 \), where \( u \) is the number of units sold. To find the number of units that maximize the profit, we can take the derivative of the profit function and set it equal to zero.

\[
P'(u) = \frac{d}{du} \left( -2u^2 + 50u - 300 \right) = -4u + 50
\]

Setting \( P'(u) = 0 \) to find the critical points:

\[
-4u + 50 = 0
\]
\[
4u = 50
\]
\[
u = \frac{50}{4} = 12.5
\]

Since \( u = 12.5 \) is a critical point, we check whether it is a maximum by evaluating the second derivative:

\[
P''(u) = \frac{d}{du} (-4u + 50) = -4
\]

Since \( P''(u) = -4 \), which is negative, the function has a maximum at \( u = 12.5 \).

Thus, the number of units that maximize the profit is \( u = 12.5 \).

### Problem 7:
You have 10 meters of string, and you need to use it to enclose the largest possible rectangle. Let the length of the rectangle be \( l \) meters and the width be \( w \) meters. The perimeter is given by:

\[
2l + 2w = 10
\]

Simplifying this:

\[
l + w = 5
\]
\[
w = 5 - l
\]

The area \( A \) of the rectangle is given by:

\[
A = l \times w = l \times (5 - l) = 5l - l^2
\]

To maximize the area, we take the derivative of \( A \):

\[
A'(l) = \frac{d}{dl} (5l - l^2) = 5 - 2l
\]

Setting \( A'(l) = 0 \):

\[
5 - 2l = 0
\]
\[
2l = 5
\]
\[
l = \frac{5}{2} = 2.5
\]

Now, substitute \( l = 2.5 \) into \( w = 5 - l \):

\[
w = 5 - 2.5 = 2.5
\]

Thus, the dimensions of the rectangle that maximize the area are \( l = 2.5 \) meters and \( w = 2.5 \) meters, forming a square.

### Problem 8:
To find the extremum of \( f(x) = x^2 + 3x - 5 \), we first compute the derivative:

\[
f'(x) = \frac{d}{dx} \left( x^2 + 3x - 5 \right) = 2x + 3
\]

Setting \( f'(x) = 0 \) to find the critical points:

\[
2x + 3 = 0
\]
\[
2x = -3
\]
\[
x = \frac{-3}{2}
\]

To determine whether this is a minimum or maximum, we evaluate the second derivative:

\[
f''(x) = \frac{d}{dx} (2x + 3) = 2
\]

Since \( f''(x) = 2 > 0 \), this critical point corresponds to a minimum. Therefore, the function has a minimum at \( x = -\frac{3}{2} \).

### Problem 9:
To find the extremum of \( f(x) = \frac{x^2 + 2x + 1}{x - 1} \), we use the quotient rule for differentiation. The quotient rule states:

\[
\frac{d}{dx} \left( \frac{g(x)}{h(x)} \right) = \frac{g'(x) h(x) - g(x) h'(x)}{[h(x)]^2}
\]

Let \( g(x) = x^2 + 2x + 1 \) and \( h(x) = x - 1 \). First, compute the derivatives:

\[
g'(x) = 2x + 2
\]
\[
h'(x) = 1
\]

Now, apply the quotient rule:

\[
f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}
\]

Simplifying the numerator:

\[
(2x + 2)(x - 1) = 2x^2 - 2x + 2x - 2 = 2x^2 - 2
\]
\[
(x^2 + 2x + 1) = x^2 + 2x + 1
\]

Now, substitute into the derivative:

\[
f'(x) = \frac{(2x^2 - 2) - (x^2 + 2x + 1)}{(x - 1)^2}
\]
\[
f'(x) = \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2}
\]
\[
f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}
\]

To find the critical points, set \( f'(x) = 0 \):

\[
x^2 - 2x - 3 = 0
\]

Solve the quadratic equation:

\[
(x - 3)(x + 1) = 0
\]

Thus, \( x = 3 \) or \( x = -1 \).

We need to check the behavior of the function near these points to determine the nature of these extrema.

**Note:** The function has a vertical asymptote at \( x = 1 \), so \( x = 1 \) is not a critical point.

To summarize:

- \( x = 3 \) is a valid critical point, and we can analyze its behavior to determine if it's a maximum or minimum.
- \( x = -1 \) is another critical point.

```python
import sympy as sp

# 6. Maximizing the profit function P(u) = -2u^2 + 50u - 300
u = sp.symbols('u')
P = -2*u**2 + 50*u - 300
P_prime = sp.diff(P, u)  # First derivative of the profit function
u_max = sp.solve(P_prime, u)  # Find the critical points

# Since we know it's a downward parabola (coefficient of u^2 is negative), this is the maximum
u_max_value = u_max[0]  # Extract the maximum point

print("Max profit is obtained at u =", u_max_value)

# 7. Enclosing the largest possible rectangle with 10 meters of string
# Let the length of the rectangle be x and the width be y, and we know 2x + 2y = 10
# We need to maximize the area A = x * y

x, y = sp.symbols('x y')
A = x * y  # Area of the rectangle
constraint = 2*x + 2*y - 10  # Perimeter constraint
y_expr = sp.solve(constraint, y)[0]  # Solve for y in terms of x

A_sub = A.subs(y, y_expr)  # Substitute y into the area function
A_prime = sp.diff(A_sub, x)  # First derivative of the area function

x_max = sp.solve(A_prime, x)  # Find the critical point for x
y_max = y_expr.subs(x, x_max[0])  # Find the corresponding y value

print(f"The dimensions of the largest rectangle are: Length = {x_max[0]} meters, Width = {y_max} meters")

# 8. Finding extremum of f(x) = x^2 + 3x - 5
f = x**2 + 3*x - 5
f_prime = sp.diff(f, x)  # First derivative of f(x)
extrema = sp.solve(f_prime, x)  # Critical points

# Second derivative test for concavity
f_double_prime = sp.diff(f_prime, x)

extrema_values = [(point, f.subs(x, point), f_double_prime.subs(x, point)) for point in extrema]
for point, value, concavity in extrema_values:
    if concavity > 0:
        print(f"Minimum at x = {point}, f(x) = {value}")
    elif concavity < 0:
        print(f"Maximum at x = {point}, f(x) = {value}")

# 9. Finding extremum of f(x) = (x^2 + 2x + 1) / (x - 1)
f2 = (x**2 + 2*x + 1) / (x - 1)
f2_prime = sp.diff(f2, x)  # First derivative of f2(x)

extrema_f2 = sp.solve(f2_prime, x)  # Critical points for f2

# Second derivative test for concavity
f2_double_prime = sp.diff(f2_prime, x)

extrema_f2_values = [(point, f2.subs(x, point), f2_double_prime.subs(x, point)) for point in extrema_f2]
for point, value, concavity in extrema_f2_values:
    if concavity > 0:
        print(f"Minimum at x = {point}, f(x) = {value}")
    elif concavity < 0:
        print(f"Maximum at x = {point}, f(x) = {value}")
```
