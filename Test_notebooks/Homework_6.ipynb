## 6. Inverse of a Matrix using the Gauss Method

To find the inverse of matrices using the Gauss method, augment each matrix with the identity matrix and perform row operations to transform the left side into the identity matrix.

---

### Matrix A:
$$
\mathbf{A} =
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}
$$

#### Step 1: Augment $\mathbf{A}$ with the identity matrix:
$$
\left[\mathbf{A} \,|\, \mathbf{I}\right] =
\begin{pmatrix}
1 & 2 & 1 & 0 \\
3 & 4 & 0 & 1
\end{pmatrix}
$$

#### Step 2: Perform row operations:
1. $R_2 \to R_2 - 3R_1$:
$$
\begin{pmatrix}
1 & 2 & 1 & 0 \\
0 & -2 & -3 & 1
\end{pmatrix}
$$

2. $R_2 \to R_2 / -2$:
$$
\begin{pmatrix}
1 & 2 & 1 & 0 \\
0 & 1 & \frac{3}{2} & -\frac{1}{2}
\end{pmatrix}
$$

3. $R_1 \to R_1 - 2R_2$:
$$
\begin{pmatrix}
1 & 0 & -2 & 1 \\
0 & 1 & \frac{3}{2} & -\frac{1}{2}
\end{pmatrix}
$$

The right-hand side is now the inverse:
$$
\mathbf{A}^{-1} =
\begin{pmatrix}
-2 & 1 \\
\frac{3}{2} & -\frac{1}{2}
\end{pmatrix}
$$

---

### Matrix B:
$$
\mathbf{B} =
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 1 \\
2 & 3 & 2
\end{pmatrix}
$$

#### Step 1: Augment $\mathbf{B}$ with the identity matrix:
$$
\left[\mathbf{B} \,|\, \mathbf{I}\right] =
\begin{pmatrix}
1 & 2 & 3 & 1 & 0 & 0 \\
4 & 5 & 1 & 0 & 1 & 0 \\
2 & 3 & 2 & 0 & 0 & 1
\end{pmatrix}
$$

#### Step 2: Perform row operations:
1. $R_2 \to R_2 - 4R_1$, $R_3 \to R_3 - 2R_1$:
$$
\begin{pmatrix}
1 & 2 & 3 & 1 & 0 & 0 \\
0 & -3 & -11 & -4 & 1 & 0 \\
0 & -1 & -4 & -2 & 0 & 1
\end{pmatrix}
$$

2. $R_2 \to R_2 / -3$:
$$
\begin{pmatrix}
1 & 2 & 3 & 1 & 0 & 0 \\
0 & 1 & \frac{11}{3} & \frac{4}{3} & -\frac{1}{3} & 0 \\
0 & -1 & -4 & -2 & 0 & 1
\end{pmatrix}
$$

3. $R_3 \to R_3 + R_2$, $R_1 \to R_1 - 2R_2$:
$$
\begin{pmatrix}
1 & 0 & -\frac{1}{3} & \frac{1}{3} & \frac{2}{3} & 0 \\
0 & 1 & \frac{11}{3} & \frac{4}{3} & -\frac{1}{3} & 0 \\
0 & 0 & -\frac{1}{3} & -\frac{2}{3} & \frac{1}{3} & 1
\end{pmatrix}
$$

4. $R_3 \to -3R_3$, $R_1 \to R_1 + R_3$, $R_2 \to R_2 - \frac{11}{3}R_3$:
$$
\begin{pmatrix}
1 & 0 & 0 & -1 & 1 & 3 \\
0 & 1 & 0 & 2 & -1 & -11 \\
0 & 0 & 1 & 2 & -1 & -3
\end{pmatrix}
$$

The right-hand side is now the inverse:
$$
\mathbf{B}^{-1} =
\begin{pmatrix}
-1 & 1 & 3 \\
2 & -1 & -11 \\
2 & -1 & -3
\end{pmatrix}
$$

---

### Matrix C:
$$
\mathbf{C} =
\begin{pmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{pmatrix}
$$

Matrix $\mathbf{C}$ is a permutation matrix. Its inverse is its transpose:
$$
\mathbf{C}^{-1} =
\begin{pmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0
\end{pmatrix}
$$

---

### Final Results:
- $$\mathbf{A}^{-1} = \begin{pmatrix} -2 & 1 \\ \frac{3}{2} & -\frac{1}{2} \end{pmatrix}$$
- $$\mathbf{B}^{-1} = \begin{pmatrix} -1 & 1 & 3 \\ 2 & -1 & -11 \\ 2 & -1 & -3 \end{pmatrix}$$
- $$\mathbf{C}^{-1} = \begin{pmatrix} 0 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 0 & 0 \end{pmatrix}$$
