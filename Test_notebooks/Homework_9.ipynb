### 1. Solve the system:
$$
\begin{cases}
x + 2y - 2z = 4, \\
2x + y + z = 0, \\
3x + 2y + z = 1
\end{cases}
$$

#### Step 1: Write the augmented matrix.
$$
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
2 & 1 & 1 & 0 \\
3 & 2 & 1 & 1
\end{array}
\right]
$$

#### Step 2: Eliminate $x$ from the second and third rows.
- Subtract $2 \times$ (row 1) from row 2:
$$
R_2 \to R_2 - 2R_1: \quad
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
3 & 2 & 1 & 1
\end{array}
\right]
$$
- Subtract $3 \times$ (row 1) from row 3:
$$
R_3 \to R_3 - 3R_1: \quad
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & -4 & 7 & -11
\end{array}
\right]
$$

#### Step 3: Eliminate $y$ from the third row.
- Subtract $\frac{4}{3} \times$ (row 2) from row 3:
$$
R_3 \to R_3 - \frac{4}{3}R_2: \quad
\left[
\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & 0 & \frac{1}{3} & \frac{1}{3}
\end{array}
\right]
$$

#### Step 4: Back-substitution.
- Solve for $z$ from row 3:
$$
z = \frac{\frac{1}{3}}{\frac{1}{3}} = 1
$$
- Substitute $z = 1$ into row 2:
$$
-3y + 5(1) = -8 \quad \Rightarrow \quad -3y = -13 \quad \Rightarrow \quad y = \frac{13}{3}
$$
- Substitute $y = \frac{13}{3}$ and $z = 1$ into row 1:
$$
x + 2\left(\frac{13}{3}\right) - 2(1) = 4 \quad \Rightarrow \quad x + \frac{26}{3} - 2 = 4
$$
$$
x = \frac{4 - \frac{26}{3} + 2}{1} = \frac{-4}{3}
$$

#### Solution:
$$
x = -\frac{4}{3}, \quad y = \frac{13}{3}, \quad z = 1
$$
