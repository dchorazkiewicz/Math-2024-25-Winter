### 1. Solve the system:
$$
3x - 2y = 5, \quad 2x + 3y = 7
$$

#### Step 1: Solve for $y$ from the first equation.
From $3x - 2y = 5$:
$$
y = \frac{3x - 5}{2}
$$

#### Step 2: Substitute $y$ into the second equation.
Substitute $y = \frac{3x - 5}{2}$ into $2x + 3y = 7$:
$$
2x + 3\left(\frac{3x - 5}{2}\right) = 7
$$
Simplify:
$$
2x + \frac{9x - 15}{2} = 7 \quad \Rightarrow \quad \frac{4x + 9x - 15}{2} = 7
$$
$$
13x - 15 = 14 \quad \Rightarrow \quad 13x = 29 \quad \Rightarrow \quad x = \frac{29}{13}
$$

#### Step 3: Find $y$.
Substitute $x = \frac{29}{13}$ into $y = \frac{3x - 5}{2}$:
$$
y = \frac{3\left(\frac{29}{13}\right) - 5}{2} = \frac{\frac{87}{13} - \frac{65}{13}}{2} = \frac{\frac{22}{13}}{2} = \frac{11}{13}
$$

#### Solution:
$$
x = \frac{29}{13}, \quad y = \frac{11}{13}
$$

### 2. Solve the system:
$$
2x - 3y = 10, \quad 4x + 5y = 20
$$

#### Step 1: Solve for $y$ from the first equation.
From $2x - 3y = 10$:
$$
y = \frac{2x - 10}{3}
$$

#### Step 2: Substitute $y$ into the second equation.
Substitute $y = \frac{2x - 10}{3}$ into $4x + 5y = 20$:
$$
4x + 5\left(\frac{2x - 10}{3}\right) = 20
$$
Simplify:
$$
4x + \frac{10x - 50}{3} = 20 \quad \Rightarrow \quad \frac{12x + 10x - 50}{3} = 20
$$
$$
22x - 50 = 60 \quad \Rightarrow \quad 22x = 110 \quad \Rightarrow \quad x = 5
$$

#### Step 3: Find $y$.
Substitute $x = 5$ into $y = \frac{2x - 10}{3}$:
$$
y = \frac{2(5) - 10}{3} = \frac{10 - 10}{3} = 0
$$

#### Solution:
$$
x = 5, \quad y = 0
$$

### 3. Solve the system:
$$
2x - y + z = 3, \quad x + 2y - z = 1, \quad 3x - y + 2z = 11
$$

#### Step 1: Solve the first equation for $z$.
From $2x - y + z = 3$:
$$
z = 3 - 2x + y
$$

#### Step 2: Substitute $z$ into the second and third equations.
Substitute $z = 3 - 2x + y$:
- Second equation: $x + 2y - (3 - 2x + y) = 1$:
  $$
  x + 2y - 3 + 2x - y = 1 \quad \Rightarrow \quad 3x + y = 4
  $$
- Third equation: $3x - y + 2(3 - 2x + y) = 11$:
  $$
  3x - y + 6 - 4x + 2y = 11 \quad \Rightarrow \quad -x + y + 6 = 11 \quad \Rightarrow \quad -x + y = 5
  $$

#### Step 3: Solve the two-variable system.
From $3x + y = 4$ and $-x + y = 5$:
1. Solve for $y$ in the first equation: $y = 4 - 3x$.
2. Substitute $y = 4 - 3x$ into $-x + y = 5$:
   $$
   -x + (4 - 3x) = 5 \quad \Rightarrow \quad -4x + 4 = 5 \quad \Rightarrow \quad -4x = 1 \quad \Rightarrow \quad x = -\frac{1}{4}
   $$

3. Find $y$: $y = 4 - 3\left(-\frac{1}{4}\right) = 4 + \frac{3}{4} = \frac{19}{4}$.

4. Find $z$: $z = 3 - 2\left(-\frac{1}{4}\right) + \frac{19}{4} = 3 + \frac{1}{2} + \frac{19}{4} = \frac{33}{4}$.

#### Solution:
$$
x = -\frac{1}{4}, \quad y = \frac{19}{4}, \quad z = \frac{33}{4}
$$
