### 1. Solve the system:
$$
\begin{cases}
   2x_1 - 3x_2 = 7, \\
   3x_1 + 5x_2 = 2
\end{cases}
$$

#### Step 1: Write the determinant of the coefficient matrix.
The coefficient matrix is:
$$
\mathbf{A} =
\begin{pmatrix}
2 & -3 \\
3 & 5
\end{pmatrix}
$$
The determinant of $\mathbf{A}$ is:
$$
\det(\mathbf{A}) =
\begin{vmatrix}
2 & -3 \\
3 & 5
\end{vmatrix}
= (2)(5) - (-3)(3) = 10 + 9 = 19
$$

#### Step 2: Solve for $x_1$ and $x_2$.
Replace the first column of $\mathbf{A}$ with the constants from the right-hand side to get $\mathbf{A}_{x_1}$:
$$
\mathbf{A}_{x_1} =
\begin{pmatrix}
7 & -3 \\
2 & 5
\end{pmatrix}
$$
$$
\det(\mathbf{A}_{x_1}) =
\begin{vmatrix}
7 & -3 \\
2 & 5
\end{vmatrix}
= (7)(5) - (-3)(2) = 35 + 6 = 41
$$
Replace the second column of $\mathbf{A}$ with the constants to get $\mathbf{A}_{x_2}$:
$$
\mathbf{A}_{x_2} =
\begin{pmatrix}
2 & 7 \\
3 & 2
\end{pmatrix}
$$
$$
\det(\mathbf{A}_{x_2}) =
\begin{vmatrix}
2 & 7 \\
3 & 2
\end{vmatrix}
= (2)(2) - (7)(3) = 4 - 21 = -17
$$

The solutions are:
$$
x_1 = \frac{\det(\mathbf{A}_{x_1})}{\det(\mathbf{A})} = \frac{41}{19}, \quad
x_2 = \frac{\det(\mathbf{A}_{x_2})}{\det(\mathbf{A})} = \frac{-17}{19}
$$

### 2. Solve the system:
$$
\begin{cases}
   2x + y - z = 1, \\
   x - y + 2z = 4, \\
   3x - 2z = -1
\end{cases}
$$

#### Step 1: Write the coefficient matrix and calculate $\det(\mathbf{A})$.
The coefficient matrix is:
$$
\mathbf{A} =
\begin{pmatrix}
2 & 1 & -1 \\
1 & -1 & 2 \\
3 & 0 & -2
\end{pmatrix}
$$
The determinant of $\mathbf{A}$ is:
$$
\det(\mathbf{A}) =
\begin{vmatrix}
2 & 1 & -1 \\
1 & -1 & 2 \\
3 & 0 & -2
\end{vmatrix}
= 2(-1)(-2) - 1(1)(-2) + (-1)(1)(3) - (-1)(-1)(3) - 2(0)(-1)
$$
$$
= 4 + 2 - 3 - 3 = 0
$$

Since $\det(\mathbf{A}) = 0$, the system cannot be solved using Cramer's rule because the matrix is singular, indicating no unique solution.

### 4. Why can't the following system be solved using Cramer's rule?
$$
\begin{cases}
x_1 + 2x_2 + 3x_3 = 3, \\
4x_1 + 5x_2 + 6x_3 = 2, \\
7x_1 + 8x_2 + 9x_3 = 1
\end{cases}
$$

#### Answer:
The coefficient matrix for this system is:
$$
\mathbf{A} =
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}
$$
The determinant of $\mathbf{A}$ is:
$$
\det(\mathbf{A}) =
\begin{vmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{vmatrix}
= 1(5 \cdot 9 - 6 \cdot 8) - 2(4 \cdot 9 - 6 \cdot 7) + 3(4 \cdot 8 - 5 \cdot 7)
$$
$$
= 1(45 - 48) - 2(36 - 42) + 3(32 - 35)
$$
$$
= -3 + 12 - 9 = 0
$$

Since $\det(\mathbf{A}) = 0$, the matrix is singular, and the system cannot be solved using Cramer's rule. This indicates that the system either has no solutions or infinitely many solutions.
