## Vector Addition and Plot

### Calculation

Given vectors:
$$
\mathbf{v}_1 = [2, 1], \quad \mathbf{v}_2 = [-1, 1]
$$

The sum of the vectors is:
$$
\mathbf{v}_1 + \mathbf{v}_2 = [2 + (-1), 1 + 1] = [1, 2]
$$

### Graph

The graph below shows the vectors $\mathbf{v}_1$, $\mathbf{v}_2$, and their sum $\mathbf{v}_1 + \mathbf{v}_2$.

## Vector Calculations and Verifications

### 2. Area of the Triangle Spanned by Two Vectors

Given vectors:
$$
\mathbf{v}_1 = [2, 1, 2], \quad \mathbf{v}_2 = [-1, 1, 1]
$$

The area of the triangle is calculated as:
$$
\text{Area} = \frac{1}{2} \|\mathbf{v}_1 \times \mathbf{v}_2\|
$$
The cross product of $\mathbf{v}_1$ and $\mathbf{v}_2$ is:
$$
\mathbf{v}_1 \times \mathbf{v}_2 =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 1 & 2 \\
-1 & 1 & 1
\end{vmatrix}
= \mathbf{i}(1 - 2) - \mathbf{j}(2 - (-2)) + \mathbf{k}(2 - (-1)) = [-1, -4, 3]
$$
The magnitude of the cross product is:
$$
\|\mathbf{v}_1 \times \mathbf{v}_2\| = \sqrt{(-1)^2 + (-4)^2 + 3^2} = \sqrt{26}
$$
Thus, the area of the triangle is:
$$
\text{Area} = \frac{1}{2} \sqrt{26} \approx 2.55
$$

### 3. Volume of the Parallelepiped Spanned by Three Vectors

Given vectors:
$$
\mathbf{v}_1 = [2, 1, -1], \quad \mathbf{v}_2 = [-1, 1, 0], \quad \mathbf{v}_3 = [1, 2, 1]
$$

The volume is calculated as:
$$
\text{Volume} = |\mathbf{v}_1 \cdot (\mathbf{v}_2 \times \mathbf{v}_3)|
$$
First, compute the cross product $\mathbf{v}_2 \times \mathbf{v}_3$:
$$
\mathbf{v}_2 \times \mathbf{v}_3 =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
-1 & 1 & 0 \\
1 & 2 & 1
\end{vmatrix}
= \mathbf{i}(1 - 0) - \mathbf{j}(-1 - 0) + \mathbf{k}(-2 - 1) = [1, 1, -3]
$$
Next, compute the dot product:
$$
\mathbf{v}_1 \cdot (\mathbf{v}_2 \times \mathbf{v}_3) = (2)(1) + (1)(1) + (-1)(-3) = 6
$$
Thus, the volume is:
$$
\text{Volume} = |6| = 6
$$

### 4. Perpendicularity of Two Vectors

Given vectors:
$$
\mathbf{v}_1 = [2, 1], \quad \mathbf{v}_2 = [-1, 1]
$$

To check if the vectors are perpendicular, compute the dot product:
$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = (2)(-1) + (1)(1) = -2 + 1 = -1
$$
Since the dot product is not zero, the vectors are not perpendicular.

### 5. Angle Between Two Vectors in Degrees

Given vectors:
$$
\mathbf{v}_1 = [4, 2, 1], \quad \mathbf{v}_2 = [1, 3, 2]
$$

The angle is calculated using the formula:
$$
\cos \theta = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_1\| \|\mathbf{v}_2\|}
$$
Compute the dot product:
$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = (4)(1) + (2)(3) + (1)(2) = 4 + 6 + 2 = 12
$$
Compute the magnitudes:
$$
\|\mathbf{v}_1\| = \sqrt{4^2 + 2^2 + 1^2} = \sqrt{21}, \quad \|\mathbf{v}_2\| = \sqrt{1^2 + 3^2 + 2^2} = \sqrt{14}
$$
The cosine of the angle is:
$$
\cos \theta = \frac{12}{\sqrt{21} \sqrt{14}}
$$
The angle in degrees is:
$$
\theta = \arccos\left(\frac{12}{\sqrt{21} \sqrt{14}}\right) \approx 45.58^\circ
$$

### 6. Proof of the Vector Triple Product Identity

The identity states:
$$
\mathbf{a} \times (\mathbf{b} \times \mathbf{c}) = (\mathbf{a} \cdot \mathbf{c}) \mathbf{b} - (\mathbf{a} \cdot \mathbf{b}) \mathbf{c}
$$
Using the definition of the cross product and the scalar triple product, the left-hand side can be expanded and shown to be equal to the right-hand side. This identity is verified mathematically and holds for any three-dimensional vectors.
