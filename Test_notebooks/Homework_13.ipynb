## Division of a Line Segment in a Given Ratio

### Problem

Divide the line segment connecting points $A(-1, 2)$ and $B(3, -2)$ in the ratio $1:3$.

### Solution

The formula for dividing a line segment in the ratio $m:n$ is:
$$
P(x, y) = \left(\frac{m x_2 + n x_1}{m + n}, \frac{m y_2 + n y_1}{m + n}\right)
$$

Here:
- $A(x_1, y_1) = (-1, 2)$,
- $B(x_2, y_2) = (3, -2)$,
- $m = 1$, $n = 3$.

Substituting the values:
$$
P(x, y) = \left(\frac{1 \cdot 3 + 3 \cdot (-1)}{1 + 3}, \frac{1 \cdot (-2) + 3 \cdot 2}{1 + 3}\right)
= \left(\frac{3 - 3}{4}, \frac{-2 + 6}{4}\right)
= \left(0, 1\right)
$$

Thus, the point dividing the line segment in the ratio $1:3$ is $P(0, 1)$.

### Graph

The graph below illustrates the points $A$, $B$, and $P$.

plt.scatter(*A, color='red', label='A(-1, 2)')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Division of Line Segment in Ratio 1:3")
plt.legend()
plt.xlim(-2, 4)
plt.ylim(-3, 3)

# Show plot
plt.show()

## Projection of a Vector onto the Axes

### Problem

Project the vector $\mathbf{a} = (3, 4)$ onto the $OX$ and $OY$ axes.

### Solution

The projection of a vector $\mathbf{a} = (x, y)$ onto the $OX$ axis is simply the vector $\mathbf{a}_x = (x, 0)$, and the projection onto the $OY$ axis is the vector $\mathbf{a}_y = (0, y)$.

Given $\mathbf{a} = (3, 4)$:

- The projection onto the $OX$ axis is:
$$
\mathbf{a}_x = (3, 0)
$$

- The projection onto the $OY$ axis is:
$$
\mathbf{a}_y = (0, 4)
$$

Thus, the projections of $\mathbf{a}$ are:
- $\mathbf{a}_x = (3, 0)$
- $\mathbf{a}_y = (0, 4)$

### Graph

The graph below illustrates the vector $\mathbf{a}$ and its projections onto the $OX$ and $OY$ axes.




````python

import matplotlib.pyplot as plt
import numpy as np

# Vector and projections
a = np.array([3, 4])
a_x = np.array([3, 0])
a_y = np.array([0, 4])

# Plot the vector and projections
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{a} = (3, 4)$')
plt.quiver(0, 0, a_x[0], a_x[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'Projection onto OX')
plt.quiver(0, 0, a_y[0], a_y[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'Projection onto OY')

# Add labels and legend
plt.text(a[0] + 0.1, a[1] + 0.1, r'$\mathbf{a} = (3, 4)$', color='r')
plt.text(a_x[0] + 0.1, a_x[1] + 0.1, r'$(3, 0)$', color='g')
plt.text(a_y[0] + 0.1, a_y[1] + 0.1, r'$(0, 4)$', color='b')

# Set limits, grid and axes
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.title("Projection of $\mathbf{a}$ onto OX and OY")
plt.legend()

# Show the plot
plt.show()
````

## Projection of a Vector onto Another Vector

### Problem

Project the vector $\mathbf{a} = (2, 3)$ onto the vector $\mathbf{b} = (1, 1)$.

### Solution

The formula for the projection of vector $\mathbf{a}$ onto vector $\mathbf{b}$ is given by:
$$
\text{proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$$

First, compute the dot products:
$$
\mathbf{a} \cdot \mathbf{b} = (2)(1) + (3)(1) = 2 + 3 = 5
$$
$$
\mathbf{b} \cdot \mathbf{b} = (1)(1) + (1)(1) = 1 + 1 = 2
$$

Now, compute the projection:
$$
\text{proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} \mathbf{b} = \frac{5}{2} (1, 1) = \left(\frac{5}{2}, \frac{5}{2}\right)
$$

Thus, the projection of $\mathbf{a}$ onto $\mathbf{b}$ is:
$$
\text{proj}_{\mathbf{b}} \mathbf{a} = \left(\frac{5}{2}, \frac{5}{2}\right)
$$

### Graph

The graph below illustrates the vectors $\mathbf{a}$, $\mathbf{b}$, and the projection of $\mathbf{a}$ onto $\mathbf{b}$.




````python

import matplotlib.pyplot as plt
import numpy as np

# Vectors and projection
a = np.array([2, 3])
b = np.array([1, 1])
proj_a_b = (np.dot(a, b) / np.dot(b, b)) * b  # Projection of a onto b

# Plot the vectors and projection
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{a} = (2, 3)$')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\mathbf{b} = (1, 1)$')
plt.quiver(0, 0, proj_a_b[0], proj_a_b[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'Projection of $\mathbf{a}$ onto $\mathbf{b}$')

# Add labels and legend
plt.text(a[0] + 0.1, a[1] + 0.1, r'$\mathbf{a} = (2, 3)$', color='r')
plt.text(b[0] + 0.1, b[1] + 0.1, r'$\mathbf{b} = (1, 1)$', color='g')
plt.text(proj_a_b[0] + 0.1, proj_a_b[1] + 0.1, r'$\left(\frac{5}{2}, \frac{5}{2}\right)$', color='b')

# Set limits, grid, and axes
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.title("Projection of $\mathbf{a}$ onto $\mathbf{b}$")
plt.legend()

# Show the plot
plt.show()
````

## Projection of a Vector onto Another Vector

### Problem

Project the vector $\mathbf{b} = (1, 1)$ onto the vector $\mathbf{a} = (2, 3)$.

### Solution

The formula for the projection of vector $\mathbf{b}$ onto vector $\mathbf{a}$ is given by:
$$
\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
$$

First, compute the dot products:
$$
\mathbf{b} \cdot \mathbf{a} = (1)(2) + (1)(3) = 2 + 3 = 5
$$
$$
\mathbf{a} \cdot \mathbf{a} = (2)(2) + (3)(3) = 4 + 9 = 13
$$

Now, compute the projection:
$$
\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{5}{13} \mathbf{a} = \frac{5}{13} (2, 3) = \left(\frac{10}{13}, \frac{15}{13}\right)
$$

Thus, the projection of $\mathbf{b}$ onto $\mathbf{a}$ is:
$$
\text{proj}_{\mathbf{a}} \mathbf{b} = \left(\frac{10}{13}, \frac{15}{13}\right)
$$

### Graph

The graph below illustrates the vectors $\mathbf{a}$, $\mathbf{b}$, and the projection of $\mathbf{b}$ onto $\mathbf{a}$.



````python
import matplotlib.pyplot as plt
import numpy as np

# Vectors and projection
a = np.array([2, 3])
b = np.array([1, 1])
proj_b_a = (np.dot(b, a) / np.dot(a, a)) * a  # Projection of b onto a

# Plot the vectors and projection
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{a} = (2, 3)$')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\mathbf{b} = (1, 1)$')
plt.quiver(0, 0, proj_b_a[0], proj_b_a[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'Projection of $\mathbf{b}$ onto $\mathbf{a}$')

# Add labels and legend
plt.text(a[0] + 0.1, a[1] + 0.1, r'$\mathbf{a} = (2, 3)$', color='r')
plt.text(b[0] + 0.1, b[1] + 0.1, r'$\mathbf{b} = (1, 1)$', color='g')
plt.text(proj_b_a[0] + 0.1, proj_b_a[1] + 0.1, r'$\left(\frac{10}{13}, \frac{15}{13}\right)$', color='b')

# Set limits, grid, and axes
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.title("Projection of $\mathbf{b}$ onto $\mathbf{a}$")
plt.legend()

# Show the plot
plt.show()
````
