### 1. Matrix Addition and Subtraction

1. $\mathbf{A} + \mathbf{B}$:
$$
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} +
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix} =
\begin{pmatrix}
6 & 8 \\
10 & 12
\end{pmatrix}.
$$

2. $\mathbf{B} - \mathbf{A}$:
$$
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix} -
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} =
\begin{pmatrix}
4 & 4 \\
4 & 4
\end{pmatrix}.
$$

3. $\mathbf{A} + \mathbf{C}$:
$$
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} +
\begin{pmatrix}
-1 & 2 \\
3 & 0
\end{pmatrix} =
\begin{pmatrix}
0 & 4 \\
6 & 4
\end{pmatrix}.
$$

4. $\mathbf{D} + \mathbf{E}$:  
**Not possible due to dimension mismatch.**

### 2. Scalar Multiplication

1. $\frac{1}{2}\mathbf{A}$:
$$
\frac{1}{2} \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} =
\begin{pmatrix}
0.5 & 1 \\
1.5 & 2
\end{pmatrix}.
$$

2. $2\mathbf{B}$:
$$

7 & 8
\end{pmatrix} =

\end{pmatrix}.
$$

### 3. Matrix Multiplication

1. $\mathbf{A} \cdot \mathbf{B}$:
$$
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} \cdot
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix} =
\begin{pmatrix}
19 & 22 \\
43 & 50
\end{pmatrix}.
$$

2. $\mathbf{B} \cdot \mathbf{A}$:
$$
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix} \cdot
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} =
\begin{pmatrix}
23 & 34 \\
31 & 46
\end{pmatrix}.
$$

3. $\mathbf{A} \cdot \mathbf{D}$:
$$
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} \cdot
\begin{pmatrix}
-1 & 2 & 3 \\
4 & 0 & 6
\end{pmatrix} =
\begin{pmatrix}
7 & 2 & 15 \\
13 & 6 & 33
\end{pmatrix}.
$$

4. $\mathbf{D} \cdot \mathbf{E}$:
$$
\begin{pmatrix}
-1 & 2 & 3 \\
4 & 0 & 6
\end{pmatrix} \cdot
\begin{pmatrix}
1 & 2 \\
4 & 5 \\
7 & 8
\end{pmatrix} =
\begin{pmatrix}
28 & 32 \\
46 & 56
\end{pmatrix}.
