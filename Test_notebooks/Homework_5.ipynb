## 1. Inverse of Matrix A

### Given:
$$
\mathbf{A} = \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}
$$

To find the inverse, compute:
$$
\mathbf{A}^{-1} = \frac{1}{\det(\mathbf{A})} \cdot \text{adj}(\mathbf{A})
$$

#### Step 1: Calculate the determinant of $\mathbf{A}$.
Using cofactor expansion along the first row:
$$
\det(\mathbf{A}) = 2 \begin{vmatrix}
1 & 0 \\
2 & 0
\end{vmatrix}
- 0 \begin{vmatrix}
0 & 0 \\
1 & 0
\end{vmatrix}
+ 1 \begin{vmatrix}
0 & 1 \\
1 & 2
\end{vmatrix}
$$

Evaluate the minors:
$$
\begin{vmatrix}
1 & 0 \\
2 & 0
\end{vmatrix} = (1)(0) - (0)(2) = 0, \quad
\begin{vmatrix}
0 & 1 \\
1 & 2
\end{vmatrix} = (0)(2) - (1)(1) = -1
$$

Substitute back:
$$
\det(\mathbf{A}) = 2(0) - 0 + 1(-1) = -1
$$

#### Step 2: Compute the adjugate of $\mathbf{A}$.
The cofactor matrix of $\mathbf{A}$ is:
$$
\text{Cof}(\mathbf{A}) = \begin{pmatrix}
\begin{vmatrix}
1 & 0 \\
2 & 0
\end{vmatrix} & \begin{vmatrix}
0 & 0 \\
1 & 0
\end{vmatrix} & \begin{vmatrix}
0 & 1 \\
1 & 2
\end{vmatrix} \\
\begin{vmatrix}
0 & 1 \\
1 & 2
\end{vmatrix} & \begin{vmatrix}
2 & 1 \\
1 & 0
\end{vmatrix} & \begin{vmatrix}
2 & 0 \\
1 & 2
\end{vmatrix} \\
\begin{vmatrix}
0 & 1 \\
1 & 0
\end{vmatrix} & \begin{vmatrix}
2 & 1 \\
0 & 0
\end{vmatrix} & \begin{vmatrix}
2 & 0 \\
0 & 1
\end{vmatrix}
\end{pmatrix}
$$

Evaluate the minors:
- Row 1: $0, 0, -1$
- Row 2: $-1, -2, 4$
- Row 3: $-1, 0, 2$

Thus:
$$
\text{Cof}(\mathbf{A}) = \begin{pmatrix}
0 & 0 & -1 \\
-1 & 2
\text{adj}(\mathbf{A}) = \text{Transpose of Cof}(\mathbf{A}) =
\begin{pmatrix}
0 & -1 & -1 \\
0 & 2 & 0 \\
-1 & 4 & 2
\end{pmatrix}
$$

#### Step 3: Calculate $\mathbf{A}^{-1}$.
$$
\mathbf{A}^{-1} = \frac{1}{\det(\mathbf{A})} \cdot \text{adj}(\mathbf{A})
$$
Substitute $\det(\mathbf{A}) = -1$ and $\text{adj}(\mathbf{A})$:
$$
\mathbf{A}^{-1} = -1 \cdot
\begin{pmatrix}
0 & -1 & -1 \\
0 & 2 & 0 \\
-1 & 4 & 2
\end{pmatrix} =
\begin{pmatrix}
0 & 1 & 1 \\
0 & -2 & 0 \\
1 & -4 & -2
\end{pmatrix}
$$

#### Verification:
Multiply $\mathbf{A}$ and $\mathbf{A}^{-1}$ to check if the product is the identity matrix:
$$
\mathbf{A} \cdot \mathbf{A}^{-1} =
\begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix} \cdot
\begin{pmatrix}
0 & 1 & 1 \\
0 & -2 & 0 \\
1 & -4 & -2
\end{pmatrix} =
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
$$
The inverse is verified.

## 2. Rank of Matrix B

### Given:
$$
\mathbf{B} =
\begin{pmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

#### Step 1: Perform row reduction to row echelon form.

1. Divide the first row by 4 to simplify the pivot:
$$
R_1 \to \frac{R_1}{4}, \quad \mathbf{B} =
\begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

2. Eliminate the first element of the second and third rows:
   - $R_2 \to R_2 + R_1$
   - $R_3 \to R_3 - 2R_1$
$$
\mathbf{B} =
\begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
0 & \frac{21}{4} & \frac{31}{4} \\
0 & \frac{39}{4} & -\frac{13}{4}
\end{pmatrix}
$$

3. Simplify the second row by dividing by $\frac{21}{4}$:
$$
R_2 \to \frac{R_2}{\frac{21}{4}}, \quad \mathbf{B} =
\begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
0 & 1 & \frac{31}{21} \\
0 & \frac{39}{4} & -\frac{13}{4}
\end{pmatrix}
$$

4. Eliminate the second element in the third row:
   - $R_3 \to R_3 - \frac{39}{4}R_2$
$$
\mathbf{B} =
\begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
0 & 1 & \frac{31}{21} \\
0 & 0 & -\frac{136}{21}
\end{pmatrix}
$$

#### Step 2: Determine the rank.
The matrix is in row echelon form, and all three rows have non-zero pivots. Thus, the rank of $\mathbf{B}$ is:
$$
\text{rank}(\mathbf{B}) = 3
$$
